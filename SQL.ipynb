### 1. Посчитайте, сколько книг вышло после 1 января 2000 года

```python
query_1 = '''SELECT COUNT (book_id)
               FROM books
               WHERE publication_date > '2000-01-01' '''
pd.io.sql.read_sql(query_1, con = engine)
```

Всего после 1 января 2000 года вышло 819 книг.

### 2. Для каждой книги посчитайте количество обзоров и среднюю оценку;

```python
query_2 = '''
SELECT b.book_id,
       b.title,
       COUNT (DISTINCT re.review_id) AS reviews,
       AVG (r.rating) AS rating_avg
FROM books AS b
LEFT JOIN reviews AS re ON b.book_id=re.book_id
LEFT JOIN ratings AS r ON b.book_id=r.book_id
GROUP BY  b.book_id
ORDER BY reviews DESC
'''
pd.io.sql.read_sql(query_2, con = engine)
```

Количество обзоров и средняя оценка представлены в столбцах reviews и rating_avg соответственно.

### 3. Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры

```python
query_3 = '''
WITH a AS(
SELECT book_id,
       title,
       publisher_id
FROM books
WHERE num_pages > 50)

SELECT p.publisher,
       COUNT (b.book_id) AS total
FROM books AS b
JOIN publishers AS p ON b.publisher_id = p.publisher_id
GROUP BY p.publisher
ORDER BY total DESC
LIMIT 1
'''

pd.io.sql.read_sql(query_3, con = engine)
```

Наибольшее количество книг с кол-вом страниц более 50 выпустило издательство Penguin Books.

### 4. Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
query_4 = '''
WITH a AS(
SELECT b.book_id,
       b.title,
       b.author_id,
       COUNT (r.rating_id) AS total_ratings
FROM books AS b
JOIN ratings AS r ON b.book_id=r.book_id
GROUP BY b.book_id)

SELECT au.author_id,
       au.author,
       AVG (r.rating) AS avg_rating
FROM a
JOIN authors AS au ON au.author_id = a.author_id
JOIN ratings AS r ON a.book_id = r.book_id
WHERE a.total_ratings >= 50
GROUP BY 1,2
ORDER BY avg_rating DESC
LIMIT 1
'''

pd.io.sql.read_sql(query_4, con = engine)
```

Автором с самой высокой средней оценкой книг, у которых более 50 оценок, является J.K. Rowling/Mary GrandPré	

### 5. Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query_5 = '''
SELECT ROUND(AVG(a.total_reviews)) AS avg_reviews
FROM (
SELECT username,
       COUNT(review_id) AS total_reviews
FROM reviews
WHERE username IN (
SELECT username
FROM ratings
GROUP BY username
HAVING COUNT(rating_id) > 50)
GROUP BY username
) AS a
'''

pd.io.sql.read_sql(query_5, con = engine)
```

Среднее количество обзоров от пользователей, которые поставили более 50 оценок, составляет 24.
